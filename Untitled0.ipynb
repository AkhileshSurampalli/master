```python
import os
import sys
import numpy as np

#mapping labels to appropriate values
labels_table = {
 
    '406516': 0,
    '406517': 1,
    '404516': 2, 
    '404517': 3,
    '406520': 4,
    '404520': 5,
    '406505': 6,
    '404505': 7,  
    '406519': 8,
    '404519': 9,
    '406511': 10,
    '404511': 11,
    '406508': 12,
    '404508': 13,
    '408512': 14,
    '407521': 15,
    '405506': 16
}

num_classes = len(labels_table) #find total number of classes (18 classes)

def shuffleInUnisson(a,b): # a and b must be vectors or arrays with the same number of lines
    assert len(a) == len(b)
    randomPermutation = np.random.permutation(len(a))
    return a[randomPermutation], b[randomPermutation], randomPermutation

def load_3Ddata_from_filelist(datafilelist,labelsfilelist,labelsTable=labels_table,permute=False):
    assert type(datafilelist) is type(labelsfilelist) #check both file list are of same type
    if isinstance(datafilelist,list):
        sortedDataList = sorted(datafilelist)
        sortedLabelsList = sorted(labelsfilelist)
        num_files = len(sortedDataList) #find number of total data files(6 files*4 subjects=24)
        assert len(sortedLabelsList) == num_files #check all data files have corresponding label files 
        
        num_windowsinfiles = np.zeros((num_files),dtype=int)  #create 1d array of 24 zeros to save number of windows in all 24 files
        
        
        for idx in range(num_files):
            labels = np.load(sortedLabelsList[idx]) #load contents of each labels file
            num_windowsinfiles[idx] = len(labels) #find length of labels = number of windows in each data/label file and save in an array
            
        total_num_windows = np.sum(num_windowsinfiles)  #total number of windows from all data/label files
        
        
        
        height=64
        width=107
        #for idx in range(1):
            #firstDataExample = np.load(sortedDataList[idx])
            #print(np.shape(firstDataExample))
            #height, width = firstDataExample.shape
            #print(height, width)
        
        #create empty 3D array for all windows in all files
        data = np.empty((total_num_windows,height,width),dtype=np.float32)
        #create empty 1D array for all window labels in all files
        labels = np.empty((total_num_windows),dtype=int)
        
        #stacking of windows as single data
        storageIdx = 0
        for idx in range(num_files):
            data_temp = np.load(sortedDataList[idx])
            labels_temp = np.load(sortedLabelsList[idx])
            data[storageIdx:storageIdx+num_windowsinfiles[idx]] = data_temp
            labels[storageIdx:storageIdx+num_windowsinfiles[idx]] = labels_temp
            storageIdx += num_windowsinfiles[idx]
            
        
    if permute:
        data, labels, permutation = shuffleInUnisson(data,labels)
    else:
        permutation = list(range(len(labels)))
        
    print("total number of windows=", total_num_windows) 
        
    #mapping all labels to class table
    for idx in range(total_num_windows):
        labels[idx] = labels_table[str(labels[idx])]
        
    
    return data, labels, data.shape, permutation
        
        
        


def load_3Ddata_filelist(filefolder,labelsTable=labels_table,permute=False):
    filelist = os.listdir(filefolder)
    
    trainingdata_filelist = [s for s in filelist if '_data' in s and ('ADL1' in s or 'ADL2' in s or 'ADL3' in s or 'Drill' in s)]
    traininglabels_filelist = [s for s in filelist if '_labels' in s and not 'last' in s and ('ADL1' in s or 'ADL2' in s or 'ADL3' in s or 'Drill' in s)]
    testingdata_filelist = [s for s in filelist if '_data' in s and ('ADL4' in s or 'ADL5' in s)]
    testinglabels_filelist = [s for s in filelist if '_labels' in s and not 'last' in s and ('ADL4' in s or 'ADL5' in s)]
    #validatedata_filelist= [s for s in filelist if ('_data' in s and 'S2' in s and ('ADL3' in s))or ('_data' in s and 'S3' in s and ('ADL3' in s))]
    #validatelabels_filelist= [s for s in filelist if ('_labels' in s and 'S2' in s and ('ADL3' in s))or ('_labels' in s and 'S3' in s and ('ADL3' in s))]
    
    
    
    
    trainingdata_list = [filefolder+'/'+s for s in trainingdata_filelist]
    traininglabels_list = [filefolder+'/'+s for s in traininglabels_filelist]
    testingdata_list = [filefolder+'/'+s for s in testingdata_filelist]
    testinglabels_list = [filefolder+'/'+s for s in testinglabels_filelist]
    #validatedata_list = [filefolder+'/'+s for s in validatedata_filelist]
    #validatelabels_list = [filefolder+'/'+s for s in validatelabels_filelist]
    
      
    
    trainingData, trainingLabels, trainingShape, trainPermutation = load_3Ddata_from_filelist(trainingdata_list,traininglabels_list,permute=permute)
    testingData, testinglabels, testingShape, testPermutation = load_3Ddata_from_filelist(testingdata_list,testinglabels_list,permute=permute)
    #validateData, validateLabels, validateShape, validatePermutation = load_3Ddata_from_filelist(validatedata_list,validatelabels_list,permute=permute)
    
    
    
    print("training shape", trainingShape)
    print("testingShape", testingShape)
    #print("validateShape", validateShape)

    return trainingData, trainingLabels, trainingShape, testingData, testinglabels, testingShape, trainPermutation, testPermutation
    #return trainingData, trainingLabels, trainingShape, testingData, testinglabels, testingShape, validateData, validateLabels, validateShape, trainPermutation, testPermutation, validatePermutation
```

```python
import keras
from keras.optimizers import RMSprop
from tensorflow.keras import models,layers
from keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization

import matplotlib.pyplot as plt
#import seaborn as sn

from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix

import time

#from data_split import *

nbSensors = 107 
timeWindow = 64

dataPath = r"/content/dataset1"
batchSize = 250
nbEpochs = 50
learningRate = 0.01 # NOTE: irrelevant if the ADADELTA optimization is used 
        
start = time.time()
x_train, y_train, trainShape, x_test, y_test, testShape, _, _ = load_3Ddata_filelist(dataPath,permute=False)
#x_train, y_train, trainShape, x_test, y_test, testShape,x_val, y_val, valShape, _,_, _ = load_3Ddata_filelist(dataPath,permute=True)
assert trainShape[1] == testShape[1] # Size of the time window
assert trainShape[2] == testShape[2] # Number of sensors

x_train = x_train.reshape(trainShape[0],trainShape[1],trainShape[2], 1)
x_test = x_test.reshape(testShape[0],testShape[1],testShape[2], 1)
inputshape = (testShape[1], testShape[2], 1)
#print('x_train shape: (%d, %d, %d)' % (trainShape[0], trainShape[1], trainShape[2]))
#print(x_test.shape)



x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
print('%d train samples' % (trainShape[0]))
print('%d test samples' % (testShape[0]))


# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)


model = models.Sequential()
model.add(BatchNormalization(input_shape=inputshape))
model.add(layers.Conv2D(64, (5, 1), activation='relu'))
model.add(layers.MaxPooling2D((2, 1)))
model.add(layers.Conv2D(64, (5, 1), activation='relu'))
model.add(layers.MaxPooling2D((2, 1)))
model.add(layers.Conv2D(64, (5, 1), activation='relu'))

model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(128, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))


model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=keras.optimizers.Adadelta(learning_rate=0.001),
                  metrics=['accuracy'],
                  #optimizer=keras.optimizers.Adagrad(lr=learningRate)
                 )

history=model.fit(x_train, y_train,
          batch_size=batchSize,
          epochs=nbEpochs,
          verbose=1,
          validation_data=(x_test, y_test)
          )

print('Model evaluation')
score = model.evaluate(x_test, y_test, verbose=0)
print("score", score)

end = time.time()

print('Total time used: %.2f seconds' % (end-start))

print(model.summary())
y_pred = model.predict(x_test)
```

```python
y_pred = model.predict(x_test)

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()

ypred=np.argmax(y_pred, axis=1)
ytest=np.argmax(y_test, axis=1)
class_names = ["Open Door 1", "Open Door 2", "Close Door 1", "Close Door 2", "Open Fridge", "Close Fridge", "Open Dishwasher", "Close Dishwasher", "Open Drawer 1", "Close Drawer 1", "Open Drawer 2", "Close Drawer 2", "Open Drawer 3", "Close Drawer 3", "Clean Table", "Drink from Cup", "Toggle Switch"]
mat = confusion_matrix(ytest, ypred)
plot_confusion_matrix(conf_mat=mat, show_normed=True, figsize=(20,20), class_names=class_names )
```
